````python
from pydantic import BaseModel, Field, EmailStr, validator
from typing import Optional, List, Dict
from datetime import datetime, date
from decimal import Decimal

# 11.1 ê¸°ë³¸ ì‚¬ìš©ì ëª¨ë¸
class User(BaseModel):
    """ì‚¬ìš©ì ê¸°ë³¸ ì •ë³´"""
    id: Optional[int] = None  # ìƒì„±ì‹œì—” ì—†ê³ , ì‘ë‹µì‹œì—” ìˆìŒ
    name: str = Field(..., min_length=2, max_length=50, description="ì‚¬ìš©ì ì´ë¦„")
    email: str = Field(..., regex=r'^[\w\.-]+@[\w\.-]+\.\w+$', description="ì´ë©”ì¼ ì£¼ì†Œ")
    age: int = Field(..., ge=0, le=150, description="ë‚˜ì´ (0-150ì„¸)")
    is_active: bool = Field(default=True, description="ê³„ì • í™œì„± ìƒíƒœ")
    created_at: Optional[datetime] = None
    
    # ì„¤ì • í´ë˜ìŠ¤
    class Config:
        # JSON ì˜ˆì‹œë¥¼ ë¬¸ì„œì— í‘œì‹œ
        schema_extra = {
            "example": {
                "name": "ì„ì´",
                "email": "seoki@example.com",
                "age": 25,
                "is_active": True
            }
        }

# 11.2 ì¤‘ì²©ëœ ëª¨ë¸
class Address(BaseModel):
    """ì£¼ì†Œ ì •ë³´"""
    street: str = Field(..., min_length=5, description="ìƒì„¸ ì£¼ì†Œ")
    city: str = Field(..., min_length=2, description="ë„ì‹œ")
    postal_code: str = Field(..., regex=r'^\d{5}$', description="ìš°í¸ë²ˆí˜¸ (5ìë¦¬)")
    country: str = Field(default="ëŒ€í•œë¯¼êµ­", description="êµ­ê°€")

class UserProfile(User):
    """ì‚¬ìš©ì ìƒì„¸ í”„ë¡œí•„ (User ëª¨ë¸ í™•ì¥)"""
    phone: Optional[str] = Field(None, regex=r'^01[0-9]-\d{4}-\d{4}$')
    address: Optional[Address] = None  # ì¤‘ì²©ëœ ëª¨ë¸
    interests: List[str] = Field(default=[], description="ê´€ì‹¬ì‚¬ ëª©ë¡")
    metadata: Dict[str, str] = Field(default={}, description="ì¶”ê°€ ì •ë³´")

# 11.3 ì»¤ìŠ¤í…€ ê²€ì¦ì (validator)
class Product(BaseModel):
    """ìƒí’ˆ ì •ë³´"""
    name: str = Field(..., min_length=2, max_length=100)
    price: Decimal = Field(..., ge=0, decimal_places=2)
    category: str
    tags: List[str] = []
    discount_percent: Optional[float] = Field(None, ge=0, le=100)
    
    @validator('name')
    def name_must_not_contain_special_chars(cls, v):
        """ìƒí’ˆëª…ì—ëŠ” íŠ¹ìˆ˜ë¬¸ì ê¸ˆì§€"""
        if any(char in v for char in ['<', '>', '&', '"', "'"]):
            raise ValueError('ìƒí’ˆëª…ì— íŠ¹ìˆ˜ë¬¸ìë¥¼ í¬í•¨í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤')
        return v.title()  # ì²« ê¸€ì ëŒ€ë¬¸ìë¡œ
    
    @validator('tags')
    def tags_must_be_unique(cls, v):
        """íƒœê·¸ ì¤‘ë³µ ì œê±°"""
        return list(set(v))  # ì¤‘ë³µ ì œê±°
    
    @validator('discount_percent')
    def discount_reasonable(cls, v):
        """í• ì¸ìœ¨ í•©ë¦¬ì„± ê²€ì¦"""
        if v is not None and v > 90:
            raise ValueError('í• ì¸ìœ¨ì´ 90%ë¥¼ ì´ˆê³¼í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤')
        return v

# API ì—”ë“œí¬ì¸íŠ¸ì—ì„œ ì‚¬ìš©
@app.post("/users", response_model=UserProfile)
async def create_user_profile(user: UserProfile):
    """ì‚¬ìš©ì í”„ë¡œí•„ ìƒì„±"""
    # IDì™€ ìƒì„±ì‹œê°„ ì¶”ê°€
    user.id = 12345
    user.created_at = datetime.now()
    
    return user  # Pydantic ëª¨ë¸ì´ ìë™ìœ¼ë¡œ JSONìœ¼ë¡œ ë³€í™˜ë¨

@app.post("/products", response_model=Product)
async def create_product(product: Product):
    """ìƒí’ˆ ìƒì„±"""
    return {
        **product.dict(),
        "id": 999,
        "created_at": datetime.now()
    }
````

````python
# 12. ì¢…í•© ì˜ˆì œ: ë¸”ë¡œê·¸ API
from typing import List, Optional
from datetime import datetime
from pydantic import BaseModel, Field
from enum import Enum

# Enum ì •ì˜
class PostStatus(str, Enum):
    DRAFT = "draft"
    PUBLISHED = "published"
    ARCHIVED = "archived"

class PostCategory(str, Enum):
    TECH = "tech"
    LIFESTYLE = "lifestyle"
    TRAVEL = "travel"
    FOOD = "food"

# Pydantic ëª¨ë¸ë“¤
class Author(BaseModel):
    """ì‘ì„±ì ì •ë³´"""
    id: int
    name: str
    email: str
    avatar_url: Optional[str] = None

class PostCreate(BaseModel):
    """ë¸”ë¡œê·¸ í¬ìŠ¤íŠ¸ ìƒì„± ìš”ì²­"""
    title: str = Field(..., min_length=5, max_length=200, description="í¬ìŠ¤íŠ¸ ì œëª©")
    content: str = Field(..., min_length=10, description="í¬ìŠ¤íŠ¸ ë‚´ìš©")
    summary: Optional[str] = Field(None, max_length=300, description="ìš”ì•½")
    category: PostCategory = Field(..., description="ì¹´í…Œê³ ë¦¬")
    tags: List[str] = Field(default=[], description="íƒœê·¸ ëª©ë¡")
    status: PostStatus = Field(default=PostStatus.DRAFT, description="ë°œí–‰ ìƒíƒœ")

class PostResponse(BaseModel):
    """ë¸”ë¡œê·¸ í¬ìŠ¤íŠ¸ ì‘ë‹µ"""
    id: int
    title: str
    content: str
    summary: Optional[str]
    category: PostCategory
    tags: List[str]
    status: PostStatus
    author: Author
    views: int = 0
    created_at: datetime
    updated_at: Optional[datetime] = None
    published_at: Optional[datetime] = None

class PostListResponse(BaseModel):
    """í¬ìŠ¤íŠ¸ ëª©ë¡ ì‘ë‹µ"""
    posts: List[PostResponse]
    total: int
    page: int
    size: int
    total_pages: int

# ê°€ìƒì˜ ë°ì´í„°
fake_author = Author(
    id=1, 
    name="ì„ì´", 
    email="seoki@blog.com",
    avatar_url="https://example.com/avatar.jpg"
)

fake_posts = [
    {
        "id": i,
        "title": f"FastAPI ë¸”ë¡œê·¸ í¬ìŠ¤íŠ¸ {i}",
        "content": f"í¬ìŠ¤íŠ¸ {i}ì˜ ìƒì„¸í•œ ë‚´ìš©ì…ë‹ˆë‹¤...",
        "summary": f"í¬ìŠ¤íŠ¸ {i} ìš”ì•½",
        "category": PostCategory.TECH,
        "tags": ["FastAPI", "Python", "API"],
        "status": PostStatus.PUBLISHED,
        "author": fake_author,
        "views": i * 10,
        "created_at": datetime.now(),
        "published_at": datetime.now()
    }
    for i in range(1, 21)  # 20ê°œ í¬ìŠ¤íŠ¸
]

# API ì—”ë“œí¬ì¸íŠ¸ë“¤
@app.get("/blog/posts", response_model=PostListResponse)
async def get_posts(
    page: int = Query(1, ge=1, description="í˜ì´ì§€ ë²ˆí˜¸"),
    size: int = Query(10, ge=1, le=100, description="í˜ì´ì§€ í¬ê¸°"),
    category: Optional[PostCategory] = Query(None, description="ì¹´í…Œê³ ë¦¬ í•„í„°"),
    status: Optional[PostStatus] = Query(None, description="ìƒíƒœ í•„í„°"),
    search: Optional[str] = Query(None, description="ì œëª© ê²€ìƒ‰")
):
    """ë¸”ë¡œê·¸ í¬ìŠ¤íŠ¸ ëª©ë¡ ì¡°íšŒ (í˜ì´ì§€ë„¤ì´ì…˜, í•„í„°ë§, ê²€ìƒ‰)"""
    
    # í•„í„°ë§ ë¡œì§
    filtered_posts = fake_posts.copy()
    
    if category:
        filtered_posts = [p for p in filtered_posts if p["category"] == category]
    
    if status:
        filtered_posts = [p for p in filtered_posts if p["status"] == status]
    
    if search:
        filtered_posts = [p for p in filtered_posts if search.lower() in p["title"].lower()]
    
    # í˜ì´ì§€ë„¤ì´ì…˜
    total = len(filtered_posts)
    start = (page - 1) * size
    end = start + size
    paginated_posts = filtered_posts[start:end]
    
    return PostListResponse(
        posts=[PostResponse(**post) for post in paginated_posts],
        total=total,
        page=page,
        size=size,
        total_pages=(total + size - 1) // size  # ì˜¬ë¦¼ ê³„ì‚°
    )

@app.get("/blog/posts/{post_id}", response_model=PostResponse)
async def get_post(post_id: int):
    """íŠ¹ì • ë¸”ë¡œê·¸ í¬ìŠ¤íŠ¸ ì¡°íšŒ"""
    # í¬ìŠ¤íŠ¸ ì°¾ê¸°
    post = next((p for p in fake_posts if p["id"] == post_id), None)
    
    if not post:
        raise HTTPException(status_code=404, detail="í¬ìŠ¤íŠ¸ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
    
    # ì¡°íšŒìˆ˜ ì¦ê°€
    post["views"] += 1
    
    return PostResponse(**post)

@app.post("/blog/posts", response_model=PostResponse, status_code=201)
async def create_post(post: PostCreate):
    """ìƒˆ ë¸”ë¡œê·¸ í¬ìŠ¤íŠ¸ ì‘ì„±"""
    new_post = {
        "id": len(fake_posts) + 1,
        **post.dict(),
        "author": fake_author,
        "views": 0,
        "created_at": datetime.now(),
        "published_at": datetime.now() if post.status == PostStatus.PUBLISHED else None
    }
    
    fake_posts.append(new_post)
    
    return PostResponse(**new_post)

@app.patch("/blog/posts/{post_id}/publish", response_model=PostResponse)
async def publish_post(post_id: int):
    """ë¸”ë¡œê·¸ í¬ìŠ¤íŠ¸ ë°œí–‰"""
    post = next((p for p in fake_posts if p["id"] == post_id), None)
    
    if not post:
        raise HTTPException(status_code=404, detail="í¬ìŠ¤íŠ¸ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
    
    post["status"] = PostStatus.PUBLISHED
    post["published_at"] = datetime.now()
    post["updated_at"] = datetime.now()
    
    return PostResponse(**post)

@app.delete("/blog/posts/{post_id}")
async def delete_post(post_id: int):
    """ë¸”ë¡œê·¸ í¬ìŠ¤íŠ¸ ì‚­ì œ"""
    global fake_posts
    
    post_index = next((i for i, p in enumerate(fake_posts) if p["id"] == post_id), None)
    
    if post_index is None:
        raise HTTPException(status_code=404, detail="í¬ìŠ¤íŠ¸ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
    
    deleted_post = fake_posts.pop(post_index)
    
    return {
        "message": "í¬ìŠ¤íŠ¸ê°€ ì„±ê³µì ìœ¼ë¡œ ì‚­ì œë˜ì—ˆìŠµë‹ˆë‹¤",
        "deleted_post_id": post_id,
        "deleted_post_title": deleted_post["title"]
    }

# í†µê³„ API
@app.get("/blog/stats")
async def get_blog_stats():
    """ë¸”ë¡œê·¸ í†µê³„ ì •ë³´"""
    total_posts = len(fake_posts)
    published_posts = len([p for p in fake_posts if p["status"] == PostStatus.PUBLISHED])
    total_views = sum(p["views"] for p in fake_posts)
    
    category_counts = {}
    for post in fake_posts:
        cat = post["category"].value
        category_counts[cat] = category_counts.get(cat, 0) + 1
    
    return {
        "total_posts": total_posts,
        "published_posts": published_posts,
        "draft_posts": total_posts - published_posts,
        "total_views": total_views,
        "average_views": total_views / total_posts if total_posts > 0 else 0,
        "category_distribution": category_counts,
        "generated_at": datetime.now()
    }
````
